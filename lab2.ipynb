```python
from PIL import Image

def get_min_max_color_values(image_path):
    # Open the image
    image = Image.open(image_path)
    # Convert the image to RGB mode
    image = image.convert('RGB')
    # Get the pixel data
    pixels = list(image.getdata())
    
    # Initialize min and max values
    min_color = (255, 255, 255)
    max_color = (0, 0, 0)
    
    # Iterate through the pixel data
    for pixel in pixels:
        min_color = tuple(min(min_color[i], pixel[i]) for i in range(3))
        max_color = tuple(max(max_color[i], pixel[i]) for i in range(3))
    
    return min_color, max_color

# Example usage
image_path = '002.jpg'
min_color, max_color = get_min_max_color_values(image_path)
print(f"Minimum color value: {min_color}")
print(f"Maximum color value: {max_color}")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def create_histogram(image_path):
    # Open the image
    image = Image.open(image_path)
    # Convert the image to RGB mode
    image = image.convert('RGB')
    # Get the pixel data
    pixels = list(image.getdata())
    
    # Separate the color channels
    reds = [pixel[0] for pixel in pixels]
    greens = [pixel[1] for pixel in pixels]
    blues = [pixel[2] for pixel in pixels]
    
    # Plot the histograms
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 3, 1)
    plt.hist(reds, bins=256, color='red', alpha=0.6)
    plt.title('Red Channel')

    plt.subplot(1, 3, 2)
    plt.hist(greens, bins=256, color='green', alpha=0.6)
    plt.title('Green Channel')

    plt.subplot(1, 3, 3)
    plt.hist(blues, bins=256, color='blue', alpha=0.6)
    plt.title('Blue Channel')
    
    plt.tight_layout()
    plt.show()

# Example usage
image_path = '002.jpg'
create_histogram(image_path)
```

```python
from PIL import Image

def convert_to_grayscale(image_path, output_path):
    # Open the image
    image = Image.open(image_path)
    # Convert the image to grayscale
    grayscale_image = image.convert('L')
    # Save the grayscale image
    grayscale_image.save(output_path)
    print(f"Image has been converted to grayscale and saved as {output_path}")

# Example usage
image_path = '002.jpg'
output_path = 'grayscale_image.jpg'
convert_to_grayscale(image_path, output_path)
```

```python
from PIL import Image, ImageFilter

def apply_filters(image_path, output_path_base):
    # Open the image
    image = Image.open(image_path)
    
    # Apply Blur filter
    blurred_image = image.filter(ImageFilter.BLUR)
    blurred_image.save(f'{output_path_base}_blurred.jpg')
    
    # Apply Contour filter
    contour_image = image.filter(ImageFilter.CONTOUR)
    contour_image.save(f'{output_path_base}_contour.jpg')
    
    # Apply Detail filter
    detail_image = image.filter(ImageFilter.DETAIL)
    detail_image.save(f'{output_path_base}_detail.jpg')
    
    # Apply Edge Enhance filter
    edge_enhance_image = image.filter(ImageFilter.EDGE_ENHANCE)
    edge_enhance_image.save(f'{output_path_base}_edge_enhance.jpg')
    
    # Apply Emboss filter
    emboss_image = image.filter(ImageFilter.EMBOSS)
    emboss_image.save(f'{output_path_base}_emboss.jpg')
    
    print("Filters applied and images saved.")

# Example usage
image_path = '002.jpg'
output_path_base = 'filtered_image'
apply_filters(image_path, output_path_base)
```
